## SETS

- Sets are unordered collection data type which are mutable ,iterable and had no duplicate elements.
- Set class represents mathematical notation of SET . These are based on a data structure known asHash table

```python
# defining an empty set
var=set()
print(var)
print(type(var))# returns class type
```

- When we assigning an empty set to an variable we use open and close braces ()
- When we declare set with some elements we use flowers brackets {}

```python
var={'Amrutha','Ravi','Sai'}
print(var)
print(type(var))
```

- Sets has no duplicates which means even if you enter duplicte elements into set will prints it without duplicates

```python
# Set with duplicates
var1={1,2,4,3,1,2,5}
print(var1)
```

```python
var2={'aswini','Balu','Pandu','aswini','pandu'}
print(var2)
```

## Dictionaries

- Dictionaries are key value pairs which are written in curly brackets '{}'and the key and values differ with colon ":"
- dictionaries are unordered collection data types which are mutable, iterable 

```python
dict={1,2,3}
type(dict)
```

```python
my_dict={'car1':"Audi",'car2':'tata','car3':'MG'}
print(my_dict)
type(my_dict)
```

```python
# indexing
print(my_dict['car1'])
```

```python
#we can loop through the dictionaries keys
for x in my_dict:
    print(x)
```

```python
# we can even loop throught the dictionaries values
for x in my_dict.values():
    print(x)
```

```python
# we can also loop throught both keys and values of  a 
for x in my_dict.items():
    print(x)
```

```python
# adding elements into dict
my_dict["car4"]='Nexon'
my_dict
```

```python
# what if give same key name to different value
my_dict['car1']='Maruthi'
my_dict
```

- it will update the value of key

#### Nested dictionaries

```python
car1_model={'Mercedes':1960}
car2_model={'Audi':1970}
car3_model={'Ambassador':1980}
car={'car1':car1_model,'car2':car2_model,'car3':car3_model}
```

```python
# Accessing elements in dictionaries
print(car['car1'])
```

```python
print(car['car2']['Audi'])
```
